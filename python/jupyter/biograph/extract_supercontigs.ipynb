```python
import biograph
```

```python
ref = biograph.Reference('/reference/hs37d5/')
```

```python
with open('hs37d5_supercontigs_nodecoy.bed', 'w') as f:
    for i in ref.supercontigs:
        if i.chromosome == 'hs37d5':
            continue
        for x in (i.chromosome, '\t', i.start, '\t', i.end, '\n'):
            f.write(str(x))
f.close()
```
